```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from urllib.request  import urlretrieve
import time
import pyautogui

service = Service("C:\\Users\\user\\Downloads\\chromedriver-win64\\chromedriver-win64\\chromedriver.exe")

prefs = {"profile.default_content_setting_values.notifications": 2}
options = Options()
# options.add_argument("headless")
options.add_argument("window-size=1920x1080")
options.add_experimental_option("prefs", prefs)
options.add_argument("--disable-gpu")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36 Edg/138.0.0.0")
options.add_argument("lang=ko_KR")

driver = webdriver.Chrome(service=service, options = options)
driver.get("https://davelee-fun.github.io/")

elements = driver.find_elements(By.CSS_SELECTOR, "div.maxthumb img")
sources = list()

for element in elements :
    sources.append(element.get_attribute("src"))

for index, source in enumerate(sources) :
    urlretrieve(source, f"IMG/{str(index)}.{source.split(".")[-1]}")
    

time.sleep(3)

driver.quit()
```

#### HTML 크롤링을 위한 XPATH 기본 포맷 (*형태)
- //tagname(@Attribute="value")
- / : 절대경로를 의미 -> ex) /html/body/div/a
- // : 문서 내 바로 검색을 하라는 의미 -> ex) //h1 (문서 내에서 바로 h1 태그 찾아오라는 의미)
- //*[@href] : 문서 내 href라는 속성을 가지고 있는 모든 태그를 다 찾아와라
- //a[@href = "https://www.gogle.com"] : 문서 내 a태그를 찾아오는데 href 속성값으로 구글탓컴을 찾아와라
- (//a)[3] : 문서 내 a태그를 찾아오는데 세번째 a만 찾아와라 (1번부터 찾아옴)
- (//table)[last()] : 문서 내 table태그를 찾아오는데 마지막 table만 찾아와라
- (//a) [position() < 3] : 문서 내 a태그를 찾아오는데 앞에서부터 2개만 찾아와라
- //div[@*] : 문서 내 div태그를 찾아오는데 어떤 속성이든 상관없이 속성이 하나라도 있으면 찾아와라
