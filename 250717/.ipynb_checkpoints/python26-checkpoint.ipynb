```python
import subprocess  # 외부 명령어 실행을 위한 모듈

url = "https://apis.data.go.kr/B551011/GoCamping/basedList?numOfRows=50&pageNo=1&MobileOS=ETC&MobileApp=camping&serviceKey=Bsmv4JkPdx5S9NKO3LG14v%2FrPvVdjEp%2BY8qSKEwLa6yiuVP7SNfSo8Ds1UbtnFDQK3Je3BVuk6cYJy6kA3hk5g%3D%3D&_type=xml"

# curl명령어를 subprocess를 통해 실행
# ["curl", "-k", url] : curl 명령으로 SSL 검증 색략하고 url요청
# capture_output=True : 표준출력(stdout)을 캡처 (파이썬에서 받아오기 위함)
# text=True : 출력 결과를 문자열로 받음
# encoding="utf-8" : 윈도우 인코딩
result = subprocess.run(["curl", "-k", url], capture_output=True, text=True, encoding="utf-8")
print(result.stdout)
```

```python
import subprocess   # curl 명령 실행하기 위해 사용
import xml.etree.ElementTree as ET # XML파싱용 문법
import requests    # 이미지다운로드용
import openpyxl    
from openpyxl.drawing.image import Image as ExcelImage    # excel 이미지 삽입용
from io import BytesIO    # 이미지 데이터를 메모리로 읽어오기 위한 객체

url = "https://apis.data.go.kr/B551011/GoCamping/basedList?numOfRows=50&pageNo=1&MobileOS=ETC&MobileApp=camping&serviceKey=Bsmv4JkPdx5S9NKO3LG14v%2FrPvVdjEp%2BY8qSKEwLa6yiuVP7SNfSo8Ds1UbtnFDQK3Je3BVuk6cYJy6kA3hk5g%3D%3D&_type=xml"

# curl 명령어로 API 요청 실행
result = subprocess.run(["curl", "-k", url], capture_output=True, text=True, encoding="utf-8")
# 응답받은 XML문자열을 ElmentTree 객체로 파싱
root = ET.fromstring(result.stdout)    

wb = openpyxl.Workbook()
ws = wb.active
ws.title = "캠핑장정보"
ws.append(["아이디", "캠핑장명", "이미지", "지역", "주소", "홈페이지"])

items = root.findall(".//item")    # xml에서 <item>태그만 출력
row = 2

for item in items :    # 각 캠핑장 정보 추출 및 엑셀에 추가
    contentId = item.findtext("contentId", default="")
    facltNm = item.findtext("facltNm", default="")
    firstImageurl = item.findtext("firstImageUrl", default="")
    doNm = item.findtext("doNm", default="")
    addr1 = item.findtext("addr1", default="")
    homepage = item.findtext("homepage", default="")

    ws.append([contentId,facltNm,"",doNm, addr1, homepage])   # 텍스트 데이터 먼저 엑셀에 입력(이미지는 나중에 입력)
    
    if firstImageurl :   # 이미지가 존재할 경우 이미지 삽입시도
        try :   # 아래 코드를 실행함 (예외가 발생할 수도 있음)
            img_data = requests.get(firstImageurl, timeout=10).content   # 이미지다운로드
            img_file = BytesIO(img_data)   # 메모리에서 파일처럼 사용
            img = ExcelImage(img_file)
            img.width = 100   # 이미지 크기 설정
            img.height = 75
            ws.add_image(img, f"C{row}")   # C열에 이미지 삽입
            ws.row_dimensions[row].height = 50   # 이미지 행 높이 조정
        except Exception as e :    # trl안에서 오류가 나면 여기로 옴, 오래내용을 e변수에 담는다
           print(f"[경고] 이미지 다운로드 실패 : {firstImageurl} -> {e}") 
    # 다음 행으로 이동        
    row += 1
    
ws.column_dimensions["B"].width = 30        
ws.column_dimensions["C"].width = 20     
ws.column_dimensions["D"].width = 15        
ws.column_dimensions["E"].width = 50        
ws.column_dimensions["F"].width = 50        

wb.save("camping_info_1.xlsx")
print("⛺camping_info_1.xlsx")
```
