1. (셀레니움을 활용해서) 알라딘 사이트에서 '자기계발' 베스트 도서의 제목과 저자를 수집하세요

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

driver.get("https://www.aladin.co.kr/shop/wbrowse.aspx?CID=336&start=we_header")

titles  = driver.find_elements(By.CSS_SELECTOR, ".b-text h4 a")
authors = driver.find_elements(By.CSS_SELECTOR, ".b-text .b-author")

for index, (title, author) in enumerate (zip(titles, authors),start=1) :
    print(f"{index}. {title.text} \n {author.text}")
    
time.sleep(1)
driver.quit()
```

2. 위키백과에서 '마케팅' 문서의 첫 문단 요약을 추출하세요.

3. 서울 열린데이터광장 공공데이터 목록에서 데이터셋 제목을 10개 수집하세요.

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

driver.get("https://data.seoul.go.kr/dataList/datasetList.do?datasetKind=1&searchFlag=M")

elements = driver.find_elements(By.CSS_SELECTOR, ".type-b a.goView")
for element in elements[:10] :
    print(element.text)
```

4. 인터파크 티켓에서 인기 콘서트 제목과 일정을 수집하세요.

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

driver.get("https://tickets.interpark.com/contents/genre/concert")

time.sleep(5)

titles = driver.find_elements(By.CSS_SELECTOR, "li.TicketItem_goodsName__Ju76j")
dates = driver.find_elements(By.CSS_SELECTOR, "li.TicketItem_playDate__5ePr2")

for idx, (title, date) in enumerate(zip(titles, dates), start=1) :
    print(f"{idx}. {title.text} - {date.text}")

driver.quit()
```
