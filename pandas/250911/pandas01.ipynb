#### pandas 라이브러리 : 전처리 & 기본 그래프 시각화 최적화 
- 파이썬에서 문서 (*파일)을 처리하는 기본·기초 지식 학습
- 특정파일 오픈 -> 파일 읽기 || 파일 쓰기 -> 파일 닫기
    - 파일 처리 내장함수 -> open(실행하고자 하는 파일경로, 파일처리 모드 옵션, 인코딩 옵션)
    - 파일처리 모드 옵션 -> r(read) / w(write) -> 기존파일에 어떤 값을 쓰기 위한 목적 / a(add) -> 새로운 데이터 추가
    - 인코딩 옵션 -> 컴퓨터를 통해서 주고 받는 문서 파일 -> 영어 -> 한글+특수문자+이모지
    - 우리의 언어로 작성한 문자열 -> 인코딩 -> 컴퓨터 -> 디코딩 -> 우리의 언어로 다시 확인
    - utf-8-sig

#### 파일 포맷 : 데이터 분야에서 가장 빈번하게 많이 사용하는 포맷
- CSV -> 엑셀 = 행 & 열 = comma-seperated values (실제 데이터 취급 시, 많이 볼 수 있는 데이터 포맷 양식)
- JSON -> 프론트 & 백엔드
- XML -> 크롤링, api

#### 250912

- xml = Extensible Markup Language = 유연한 마크업 언어로 되어있는 데이터 포맷 양식
- json => JavaScript Object Notation = 클라이언트와 서버가 데이터를 서로 주고 받을 때 사용하는 데이터 포맷 양식
  - key-value 쌍으로 이루어진 구조
  - 문자열 형식으로 전달되지만 구조화된 데이터
  - 파이썬에서는 json 데이터를 딕셔너리로 다룸
  - json.loads() : 문자열 → dict
  - json.dumps() : dict → 문자열
  - json.load() : 파일 → dict
  - json.dump() : dict → 파일
