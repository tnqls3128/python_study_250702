#### 1. pandas
- pandas는 파이썬에서 사용할 수 있는 라이브러리 (VS 프레임워크)
- 행과 열로 구성되어있는 테이블 형태의 데이터를 다룰 때 효과적으로 관리해줄 수 있는 라이브러리
- 왜 pandas?
- 1) 가장 오랜 기간 테이블 형태의 데이터를 다룰 때 사용 -> 검증 -> 버그가 거의 수정 & 안정
  2) 현업에서 가장 많이 사용
  3) 기능이 압도적 -> plotly 데이터 시각화 라이브러리와 호환성 -> 왠만한 전처리 & 데이터 시각화
- pandas를 통해서 실행하고자 하는 기본 기능 위주로 문법 기초 학습
- 학습된 결과를 토대로 2가지 실습 (코로나 확진자 데이터를 전처리·시각화 실습 // 이커머스 쇼핑몰 데이터 전처리 & 시각화)
- raw data : raw => 원본 데이터 기준 전처리
- EDA : 탐색화 데이터 분석 => Exploratory Data Analysis
- 1) 해당 데이터는 어떤 출처
  2) 해당 데이터의 양은 어느정도인가
  3) 해당 데이터 내 각 feature("하나의 field, 컬럼값") 상관관계는 어떤 것인가

#### 2. pandas를 학습할 때 중요하게 생각해야하는 개념
- Series : 1차원 데이터 => 1개의 필드명 + 컬럼 한 줄 
- Dataframe : 2차원 데이터  => 행과 열로 구성되어있는 테이블형 데이터
- 0, 1, 2 => 인덱스 (*좌측 기준)
- 미국, 한국, 중국 => "필드명/값" (*상단 기준)
- 2.1, 2.2, 2.3 => 각각의 필드명에 속해있는 요소들의 집합 => "컬럼"
- 2.1, 2.2, 2.3, 10, 13, 15 => values => "밸류즈"
- CRUD (*create / read / update / delete)
- pandas 데이터 타입
  1) int64 : 정수
  2) float64 : 실수
  3) bool : 논리형
  4) object : 문자열, 1개의 시리즈 안에 복수의 데이터 
  5) datetile64 : 날짜, 시간 타입

#### 3. pandas > DataFrame 개념 정의
- 행과 열로 구성되어 있는 테이블형 데이터 = 2차원 데이터
- 시리즈 = 1차원 데이터 => 필드명 : [1,2,3]
- 세로 방향의 1차원 데이터 확인, 분석
- 가로 방향의 1차원 데이터 확인, 분석

#### 4. 탐색적 데이터 분석 과정을 통한 데이터 전처리
- EDA : Exploratory Data Analysis // 미국 통계학자 존 튜키
- 개념 및 이론 : 데이터 분석을 위해서 raw data를 다양한 각도에서 관찰하고 해당 데이터를 이해하는 과정
  - 하지만 데이터의 인사이트를 도출하거나 이해하는 정답은 존재하지 않는다는 것은 기본 전제
  - 그러나 수많은 데이터 분석가들이 데이터를 읽으면서 내린 공통적인 결론 및 의견들을 조합해서 하나의 이론 => EDA
    1) 데이터의 출처와 주제에 대한 이해
    2) 데이터 크기 확인
    3) 데이터 구성요소(feature)의 속성(특징)확인
