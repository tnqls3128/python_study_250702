```python
import csv

with open("00_data/tmp_csv.csv", "w", encoding="utf-8-sig", newline="") as writter_csv :
    field_name_list=["First Name", "Last Name"]
    writer = csv.DictWriter(writter_csv, fieldnames=field_name_list)
    writer.writeheader()  # 위 문장에서 선언한 필드명을 실제 csv 파일에 적용하겠다라는 함수
    writer.writerow({"First Name" : "Dave", "Last Name" : "Lee"})  # 두번째 행
    writer.writerow({"First Name" : "David", "Last Name" : "Park"})  
    writer.writerow({"First Name" : "Robin", "Last Name" : "Kim"})
```

```python
import csv

with open("00_data/tmp_csv.csv", "r", encoding="utf-8-sig", newline="") as reader_csv :
    reader = csv.DictReader(reader_csv)  # 리스트 형식으로 나옴
    for row in reader :
        print(row["First Name"], row["Last Name"])
```

```python
from bs4 import BeautifulSoup

data_file = open("00_data/users.xml", "r", encoding="utf-8-sig")
soup = BeautifulSoup(data_file, "xml")

users = soup.select("user")  # 해당 문서 안에 동일한 태그를 모두 찾아와서 리스트 형태의 타입
# users = soup.select_one("user")  # 해당 문서 안에 가장 첫번째로 검색된 태그만 찾아와서 리스트형태의 타입으로 변환

for user in users :
    print(f"이름 : {user.select_one("name").text}")
```

```python
import requests  # 크롤러
from bs4 import BeautifulSoup

keyword = "ai"
google_related_keyword_api = f"http://suggestqueries.google.com/complete/search?output=toolbar&q={keyword}"
response = requests.get(google_related_keyword_api)
soup = BeautifulSoup(response.content, "xml")   # 외부 웹페이지 주소 content 값이므로  

dataxml = soup.select("suggestion")

for item in dataxml :
    print(item["data"])
```

```python
import json

data = '{"id" : "01","language" : "python","edition" : "third","author" : "David"}'

jsondata = json.loads(data)  # json 형식의 문자열을 받아서 파이썬 딕셔너리로 변환
# 파이썬 셀 혹은 에디터를 통해서 직접 문자열 형태의 제이선 문서를 로드해오고자 할 때
print(jsondata["id"],jsondata["language"],jsondata["edition"], type(jsondata))

# json.loads => 문자열을 dic으로 반환

#01. 최초에 웹 이라는 생태계가 탄생 -? JS
#02. 클라이언트 & 서버의 통신 -> JSON
#02. 딕셔너리라는 자료구조 
```

```python
import json

data = {"id" : "01","language" : {"python" : "basic", "java" : "advanced"},"edition" : "third","author" : "David"}

jsondata = json.dumps(data)  # dumps함수는 loads와 달리 로컬에서 가져오고자 하는 json 값에 문자열 처리를 굳이 하지 않아도 됨
print(jsondata, type(jsondata))

jsondata02 = json.dumps(data, indent=2)
print(jsondata02, type(jsondata02))

# json.dumps() => dict을 문자열로 반환 
```
