```python
# 데이터프레임에서 중복 행 확인/제거

import pandas as pd
PATH = "COVID-19-master/csse_covid_19_data/UID_ISO_FIPS_LookUp_Table.csv"
```

```python
doc = pd.read_csv(PATH, encoding = "utf-8-sig")
doc = doc[["iso2", "Country_Region"]]
doc
```

```python
doc.duplicated()  #중복되어있는지 아닌지 알 수 있음
# 현재 여러번 중복해서 사용된 값을 1개만 남기고 나머지는 제거
# 어떤 값을 살리고, 어떤 기준으로 나머지 값을 제거할건지?
```

```python
doc = doc.drop_duplicates(subset = "Country_Region", keep="last")  # 중복된 값 제거, 그 중 마지막에 나오는 행만 남기고 나머지는 제거
doc
```
