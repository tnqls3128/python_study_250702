```python
# 함수 선언"문" 시작!!!
def add(a, b) :
    return a + b

# 함수 호출!!!
print(add(1, 2))

print(add(4, 5))

print(add(9, 3))
```

```python
# 함수 => 람다"식" => 익명함수
# 함수의 이름을 없애는 대신, 해당 함수를 여튼 어떠한 변수 안에는 넣어줘야함

add = lambda a, b : a + b

print(add(1, 2))
```
