```python
#   = Multiple Line

import plotly.graph_objects as go
import numpy as np
import pandas as pd

df = pd.DataFrame(
    {
        "A" : np.random.rand(50),
        "B" : np.random.rand(50),
    }
)

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=df.index,
        y=df["A"],
        mode="lines",
        name="A",
        showlegend=True
    )
)

fig.add_trace(
    go.Scatter(
        x=df.index,
        y=df["B"],
        mode="lines+markers+text",
        name="B",
        showlegend=True,
        text=df.index,
        textposition="top center" #    
    )
)

fig.show()
```
