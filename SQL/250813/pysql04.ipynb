```python
# https://davelee-fun.github.io/
# https://davelee-fun.github.io/page2/
# https://davelee-fun.github.io/page3/

import requests
from bs4 import BeautifulSoup
import pymysql

db = pymysql.connect(
    host = "localhost",
    port = 3306,
    user = "root",
    passwd = "753951DnDn!!",
    db = "ecommerce_v2",
    charset = "utf8"
)

cursor = db.cursor()


for page_num in range(10) :
    if page_num == 0 :
        res = requests.get("https://davelee-fun.github.io/")
    else :
        res = requests.get(f"https://davelee-fun.github.io/page{str(page_num + 1)}/")
    soup = BeautifulSoup(res.content, "html.parser")
    data = soup.select("div.card-body")

    for item in data :
        category = item.select_one("h2.card-title").get_text().replace("관련 상품 추천", "").strip()
        product = item.select_one("h4.card-text").get_text().replace("상품명", "").strip()
        print(category, product)

        # sql = f"INSERT INTO teddyproducts (TITLE, CATEGORY) VALUES({product}, {category})"
        sql = "INSERT INTO teddyproducts (TITLE, CATEGORY) VALUES(%s, %s)"
        cursor.execute(sql, (product, category))
        # cursor.execute(INSERT INTO teddyproducts (TITLE, CATEGORY) VALUES(%s, %s)", (product, category)")
        
db.commit()
db.close()
```
