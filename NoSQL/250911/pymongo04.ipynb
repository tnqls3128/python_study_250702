```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017")
db = client.sample_mflix
movies = db.movies

print(movies)
```

```python
# "star" 라는 단어가 포함된 영화제목을 가져오세요
import re

# regax = re.compile("star", re.I)
# # re.I => 대소문자를 구분하지 않겠다는 옵션
# # re.compile => 정규표현식 객체로 만들어줌

# for movie in movies.find({"title" : regax}).limit(3) : 
#     print(movie)

star_movies = movies.find({"title" : {"$regex" : ".*Star.*"}}, {"title" : 1, "_id" : 0})

for movie in star_movies :
    print("영화제목 :", movie["title"])
```

```python
# 각 감독들이 제작한 영화의 수가 25개 이상인 감독들만 찾아서 출력해주세요
pipelines = [
    {"$unwind" : "$directors"},
    {"$group" : {"_id" : "$directors", "count" : {"$sum" : 1}}},
    {"$match" : {"count" : {"$gte" : 25}}}
]

directors_count = list(movies.aggregate(pipelines))

for  director in directors_count :
    print(f"감독 : {director["_id"]}, 영화수 : {director["count"]}편")
```

```python
# 평점을 기준으로 상위 5개 영화를 찾으세요 (단 투표를 1000표 이상 받은 영화에 한함)
top_rated_movies = list(movies.find({"imdb.votes" : {"$gte" : 1000}}).sort("imdb.rating", -1).limit(5))

for movie in top_rated_movies :
    print(f"영화제목 : {movie["title"]}, 평점 : {movie["imdb"]["rating"]}")
```
