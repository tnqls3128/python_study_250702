#### 1. 데이터베이스 = DB
- 데이터베이스 => 데이터를 보관하고 저장할 수 있는 큰 공간
- 데이터베이스 역사 : 오래되었음
- 데이터를 담고 관리할 수 있는 시스템, 체계의 역사 : 오래되었음
- DBMS : 오래되었음
- 80여개 정도의 DBMS
- 관계형 데이터베이스 관리 시스템 : RDBMS
- Oracle > MySQL
- SQL -> RDBMS 기반하에 저장, 보관되어있는 데이터를 처리하기 위한 목적의 언어
- 스마트폰 -> 모바일 -> 데이터 기하급수적 증가
- 빅데이터 하루에도 수백,수천만건의 데이터가 쌓여지는 것을 의미
- RDBMS -> DB -> Table -> Schema
- 빅데이터라는 방대한 데이터의 입·출력 건이 수시로 실행되고 있는데 이것을 과거의 문법 및 방법으로 처리·관리한다는 것이 현실/물리적으로 부담 시작
- 더이상 SQL문법만 가지고 RDBMS 라는 데이터 관리 시스템만 의존하는 것은 구시대적인 사고방식
- NoSQL => Not Only SQL
- NoSQL기반 데이터 처리 -> 전통적인 역사를 가지고 있는 기업X  ex. 토스
- NoSQL => MongoDB

#### 2. NoSQL의 이해
- NoSQL은 스키마 & JOIN 등의 문법 및 관리 시스템이 존재하지 않음
- 현재 작업하는 업무가 95%가 read, 5% write => RDBMS (SQL)
- 현재 작업하는 업무가 50%이상 write => NoSQL + SQL

#### 8. MongoDB Aggregation Framework
- MongoDB -> No SQL, Aggregation -> 집합 | 집계
- Aggregation -> 프레임워크
- \$push <-> 배열이 아닌 요소들을 하나의 배열의 자료구조 형태로 만들어주는 기능
- 프레임워크 | 라이브러리 
- 기존에 학습했던 find()의 상위호환 버전이라고 생각해도 무방
- Linux운영체제 -> pipeline 개념을 벤치마킹 -> 함수의 기능이 구현 -> 기능
- shard(독립적인 각각의 요소) : 샤드 -> A (*1번 샤드) -> B (*2번 샤드)  
- Aggreagtion Framework를 실행하기 위한 주요 단계
  - \$addFields : 기존 필드에 추가적으로 속성 더하거나 변경하고자 할 때
  - \$match : 조건에 맞는 문서를 필터링
  - \$unwind : 종합배열의 값을 하나로 합쳐주는 역할
  - \$group : 문서를 그룹화하고 집계 작업을 수행할 수 있도록 준비
  - \$project : 연산처리된 결과를 특정 요소에 포함시키거나 새로운 문서를 변환 작업
  - \$limit : 특정 개수의 문서를 출력하고자 할 때
---
  - \$sort : 연산처리된 결과를 정렬 (*위치 무관)
  - \$skip : 특정 연산처리 후 의도적으로 문서를 건너뛰고 값을 찾아오려고 할 때
  - \$count : 연산결과값을 통해서 받아온 문서 수의 개수를 집계
  - \$lookup : 서로 다른 컬렉션을 묶어서 사용할 때 사용 (*JOIN)
  - \$document : 임시 메모리 공간에 문서를 생성해놓을 때 사용
  - \$merge : 임시 메모리 공간에 저장된 문서를 특정 컬렉션 안에 삽입
    (*into : 삽입할 컬렉션 명)
  - \$facet : 여러개 파이프라인을 동시에 실행 , 어떤 하나의 측면 => interface : aggregation 프레임워크를 사용할 때 단점을 보완할 수 있음 => 파이프라인 => 앞에서 설정해놓은 단계가 종료되어야 그 다음 단계가 실행
    (*만약, match 연산을 통해서 찾아와야하는 조건식이 100줄 // 이 후 group 3줄 => 파이프라인 내 상호간 순서에 영향을 받지 않는 요소라고 한다면, 동시 실행을 시킬 수 있음, 앞쪽에서 실행되어야할 요소를 기다리지 않고 병렬로 같이 실행) => 그룹을 생성, A그룹 & B그룹 (영향받지않으므로 가능)
   - \$redact : 사전적의미 => "수정하다" (*어떤 조건에 따라서 수정 여부를 판단, cond랑 같이 씀)
   - \$cond : 조건을 따져오는 역할
     (if / then / else)
---

    (*연산처리를 하기 위한 집계 작업 명령어 -> 예약어)
      - \$sum : 특정 필드명 안에 포함되어있는 값의 합계
      - \$avg : 특정 필드명 안에 포함되어있는 값의 평균
      - \$min : 특정 필드명 안에 포함되어있는 최소값
      - \$max : 특정 필드명 안에 포함되어있는 최대값
      - \$first : 특정 필드 안에 첫번째 값
      - \$last : 특정 필드 안에 마지막번째 값
      - \$push : 기존값들을 배열의 형태로 구성할 때 사용
      - \$convert : 특정 필드 내 속성을 변환시킬 때 사용
      - \$strLenCP : 문자열의 길이 값을 찾아올 때 사용
      - \$toString : 문자열 형태가 아닌 타입의 값을 문자로 변환
      - \$addToSet : 특정 필드 안에 있는 값을 중복하지 않고 찾아오고자 할 때
      - \$concat : 각 필드 안에 존재하는 값을 하나의 문자열로 생성하고자 할 때

* 필드명 작성 순서
1. \$match, $group : 해당 명령어 안에 어떤 조건식을 통해서 결과값이 도출되었음, 그리고 해당 결과 값을 어떤 필드에 값을 담아야 하는 경우 -> 무조건 내부적으로 필드명을 먼저 사용!
2. \$count, \$sum, \$avg : 어떤 값들을 집계한 후, 새로운 필드를 생성할 때 -> 무조건 예약어(*명령어)를 먼저 사용 후 이후에 속성값으로 필드명을 입력

#### 9. SQL JOIN VS NoSQL lookup
1) SQL => ROBMS (* 관게형 테이터베이스 관리 시스템)
   - schema 개념 // 테이블을 생성, 필드를 선언 => Promary key, Forign key => index
   - INNER JOIN : 서로 다른 테이블에서 교집합 (* 매칭되는 행만 가져옴)
   - LEFT OUTER JOIN : 메인이 되는 왼쪽 테이블을 모두 가져오고 서브가 되는 오른쪽 테이블은 매칭 되는 값만 찾아옴
   - RIGHT OUTER JOIN : LEFT OUTER JOIN 반대개념
2) NoSQL => 기본적으로 무조건 LEFT OUTER JOIN의 형태를 가지고 있다

```
db.comments.arregate([
    {
        $loockup :
            {
                from : "movies"  // 조인할 대상의 컬렉션 이름
                localField : "movie_id"  // 현재 컬렉션(comments)에서 조인에 사용할 필드
                foreignField : "_id"  // 조인할 컬렉션에서 조인에 사용할 필드
                as : "movie"  // 조인된 결과를 저장할 필드 이름
            }
    },
    {
        $match : {customerInfo : {$ne : []}}
    }
])
```
