```python
from pymongo import MongoClient
client = MongoClient("mongodb://localhost:27017")

db = client.sample_mflix
comments = db.comments
```

```python
pipeline = [
    {"$addFields" : {"textLength" : {"$strLenCP" : {"$ifNull" : ["$text", ""]}}}},
    {"$group" : {"_id" : "$email", "commentsCount" : {"$sum" : 1}, "avgTextLen" : {"$avg" : "$textLength"} , "lastCommentDate" : {"$max" : "$date"}}},
    {"$match" : {"commentsCount" : {"$gte" : 10}}},
    {"$sort" : {"commentsCount" : -1, "avgTextLen" : -1, "lastCommentDate" : -1}}
]

for comment in comments.aggregate(pipeline):
    print(f'이메일 : {comment["_id"]}, 댓글 수 : {comment["commentsCount"]}, 댓글 평균길이 : {comment["avgTextLen"]:.2f}, 최근 댓글 시점 : {comment["lastCommentDate"]}')
```
