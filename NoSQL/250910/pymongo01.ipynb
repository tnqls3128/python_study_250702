```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017")

db= client["test"]
collection = db.users
data = {"name": "Jonn", "age": 30}
result = collection.insert_one(data)
print("Inserted ID : ", result.inserted_id)

data = [
    {"name": "Alice", "age": 25},
    {"name": "Bob", "age": 35},
    {"name": "Charie", "age": 40}
]

result = collection.insert_many(data)
print("Inserted IDs : ", result.inserted_ids)
```

#### find() = find_one()

#### 데이터 문서 수정
1. update_one() / update_many()
   => collection.update_one("조건 = filter", "변경할 값 = update")
2. delete_one() / delete_many()

```python
collection = db.users

filter = {"name" : "Jonn"}
update = {"$set" : {"age" : 31}}
result = collection.update_one(filter, update)

print("Modified Count : ", result.modified_count)
```

```python
filter = {"age": {"$gte": 40}}
update = {"$set": {"is_available": True}}
result = collection.update_many(filter, update)

print("Modified Count:", result.modified_count)
```

```python
collection = db.users
filter = {"name": "Jonn"}
result = collection.delete_one(filter)

print("Deleted Count:", result.deleted_count)
```

```python
collection = db.users
filter = {"age" : {"$lt" : 10}}
result = collection.delete_many(filter)

print("Deleted Count:", result.deleted_count)
```

```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017")
db = client.sample_mflix
movies = db.movies

print(type(movies))
```

```python
# JS파일 포맷에서는 객체안에 있는 값을 찾아올 때, 대괄호&온점 표기법 가능
# ㅠㅏ이썬 포맷에서는 객체형태를 띄고있는 자료구조를 딕셔너리 => 대괄호 표기법으로 특정 키 내부의 값을 찾아온다

for movie in movies.find({"year" : 1923}, {"_id" :0, "title":1, "year":1}) :
    print(movie, movie["year"])
```

```python
# MongoDB 안에 있는 특정 DB의 컬렉션 값을 어떤 조건과 함께 찾아오게 되면 기본적으로 리스트 형태를 띄고 있음
# 해당 리스트 자료구조 안에있는 각각의 문서는 딕셔너리의 자료구조로 되어있다

for movie in movies.find(
    {"$or" : [{"year" : {"$lt" : 1900}}, {"year" : {"$gt" : 2015}}]},
    {"_id" : 0, "title" : 1, "year" :1}
) :
    print(movie)
```

```python
# $all : 배열 내부에 있는 값이 조건으로 모두를 충족시키게끔

for movie in movies.find(
    {"genres" : {"$all" : ["Action", "Sci-Fi", "Drama"]}},
    {"_id" : 0, "title" : 1, "year" : 1, "genres" : 1}
):
    print(movie)
```

```python
for movie in movies.find().sort("imdb.rating", -1).skip(3).limit(3):
    print(movie)
```

```python
# 정규표현식

import re

regax = re.compile("Star", re.I)

for movie in movies.find({"title" : regax}).limit(3) :
    print(movie)
```

```python
for movie in movies.find({"title" : {"$regex" : "Star", "$options" : "i"}}).limit(1) :
    print(movie)
```

```python
distinct_genres = movies.distinct("genres")
print(distinct_genres)
```

```python
# $in : 배열 내 필드 값 중 하나만이라도 가지고 있다면 찾아올 수 있음

for movie in movies.find({"genres" : {"$in" : ["Action", "Adventure"]}}).limit(3) : 
    print(movie)
```

```python
#  $exist : 존재하다 => 특정 필드가 문서에 존재하는지 여부에 따라서 문서를 선택해서 값을 찾아엄

for movie in movies.find({"writers" : {"$exists" : False}}).limit(3) :
    print(movie)
```

```python
count = movies.count_documents({"genres": {"$in" : ["Action", "Adventure"]}})
print(count)
```
