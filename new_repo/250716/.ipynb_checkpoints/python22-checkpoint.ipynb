```python
import requests    # html웹 요청보내는 라이브러리
import pprint    # 자료 보기좋게 정렬
import openpyxl    # 엑셀파일 생성하고 다룸

client_id = "hzjIdWk02iJtdRzybg2f"
client_secret = "geGo4UBMp9"
start,num = 1,0    # start는 검색시작위치, num은 출력 순서번호 누적하기 위한 변수

excel_file = openpyxl.Workbook()    # 새 엑셀 파일 생성
excel_sheet = excel_file.active    # 기본시트 활성화
excel_sheet.column_dimensions["B"].width = 100    #열 너비 조절
excel_sheet.column_dimensions["C"].width = 100
excel_sheet.append(["No", "제목", "링크"])    # 첫번째 행 추가

for index in range(10) :    # 10번 반복(한번에 100개만 가져올 수 있어서 1000개를 가져오기 위함)
    start_num = start + (index * 100)   
    naver_open_api = "https://openapi.naver.com/v1/search/shop.json?query=무신사&display=100&start=" + str(start_num)
    header_params = {"X-Naver-Client-Id":client_id, "X-Naver-Client-Secret":client_secret} 
    
    res = requests.get(naver_open_api, headers=header_params)    
    
    # test = res.json()
    # test
    
    if res.status_code == 200 :
        data = res.json()
        # pprint.pprint(data)
        # print(data["items"][0]["title"])  -> 첫번째 상품 제목
        # for index, item in enumerate(data["items"]) :   -> 상품나열하는 코드(출력번호 누적X)
            # print(index + 1, item["title"], item["link"])
        for item in data["items"] :
            num+=1
            excel_sheet.append([num, item["title"], item["link"]])    #엑셀에 한줄 추가
            print(num, item["title"], item["link"])    #콘솔에도 출력
             
    else :
        print("Error Code: ",res.status_code)

excel_file.save("shopping_with_무신사.xlsx")
excel_file.close()
```

```python
#실습

import json
import requests
import pprint
import openpyxl

client_id = "hzjIdWk02iJtdRzybg2f"
client_secret = "geGo4UBMp9"
start, num = 1,0

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.append(["No", "제목", "링크"])


for index in range(10) :
    start_num = start + (index * 100)  
    naver_open_api = "https://openapi.naver.com/v1/search/blog.json?query=롯데월드&display=100&start="+ str(start_num)
    header_params = {"X-Naver-Client-Id" : client_id, "X-Naver-Client-Secret" : client_secret}

    res = requests.get(naver_open_api, headers = header_params)

    if res.status_code == 200 :
        data = res.json()
        for item in data["items"] :
            num+=1
            excel_sheet.append([num, item["title"], item["link"]])
            print(num, item["title"], item["link"]) 

    else :
        print("Error Code: ",res.status_code)

excel_file.save("blog_with_롯데월드.xlsx")
excel_file.close()
```

```python
#실습예제
import json
import requests
import openpyxl

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.column_dimensions["B"].width = 100
excel_sheet.column_dimensions["C"].width = 100
excel_sheet.append(["제목", "개봉일", "평점", "줄거리"])

url = "https://api.themoviedb.org/3/movie/now_playing?api_key=48fad6c19d4f9059650d2f5fc123766e&language=ko-KR&page=1"

res = requests.get(url)

if res.status_code == 200 :
    data = res.json()
    num=0
    for item in data["results"] :
       num+=1
       excel_sheet.append([num, item["title"], item["release_date"], item["vote_average"], item["overview"]])
       print(num, item["title"], item["release_date"], item["vote_average"], item["overview"])

excel_file.save("movie_width_수빈tmdb.xlsx")
excel_file.close()
```
