#### 상자그림 그래프
1) Box Plot (*Plot => Plotting = 그리다)
2) 하나의 전체 데이터 내부에 소속되어있는 내부값들의 데이터 분포도 확인 => 데이터 분포의 주요 지표값을 확인
3) Box Plot의 데이터 분석, 대전제 -> 반드시 전체 데이터를 4등분 = 4분위로 쪼개서 분석
4) 최소값 ~25% | 26%~50% | 51%~75% | 76%~100%
5) 전체 데이터를 구성하고 있는 각각의 요소들을 어떻게 지칭할 것인가?
    - Min (*최소값)
    - Q1 (*1사분위수 : 하위 25%)
    - 중앙값
    - Q3 (*3사분위수 : 상위 25%)
    - 최대값
    - IQR (*Interquartile Range) -> 1사분위와 3사분위 사이의 있는 값
    - 이상치
      - Upper Fence (*상한 이상치)
      - Lower Fence (*하한 이상치)
  

- data = [1,2,3,100] => 2.5
- data = [1,2,3,6,100] => 6
- => 중간값 : 50, 3
- => 후보1 = 53
- => 후보2 = 1 + 100 / 2 = 50
    
    - 해당 데이터 내 세부값의 개수가 홀수인 경우에는 정 가운데 값이 중앙값
    - 데이터의 세부값이 짝수인 경우에는 해당 데이터의 정 가운데를 기준으로 좌, 우측 값을 가져와서 평균값 2.5

- data = [3,5,7,8,12,13,14,18,20,21]
  - min : 3
  - max : 21
  - median : (12+13)/2
  - Q1(25%) : 6.5
  - Q3(75%) : 18.5
  - IQR : 12 = 18.5 - 6.5
  - Lower Fence : -11.5 = 6.5 - (1.5 * 12)
  - Upper Fence : 36.5 = 18.5 + (1.5 * 12)
- 1사분위 & 3사분위 계산하는 공식
  - 2가지 방식 계산 공식  # n = 10 => 하나의 선에서 길게 늘어뜨린 공간안에 배치된 값들의 연속, n+1
  1) (n+1)*0.25 = (10+1)*0.25 => 2.75번째 (*1사분위)
     > 2번째 = 5 / 3번째 = 7
     > 5 + (7-5) * 0.75 = 6.5
     > (n-1) * 0.25
  3) (n+1)*0.75 = (10+1)*0.75 => 8.25번째
     > 8번째 = 18 / 9번째 = 20
     > 18 + (20-18) * 0.25 = 18.5
     > (n-1) * 0.75
